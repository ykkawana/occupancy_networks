```python
import trimesh
import numpy as np
import kaolin as kal
from kaolin.transforms import meshfunc
import torch
```

```python
def rotz(d):
    d = d / 180. * np.pi
    return np.array([
        [1., 0., 0.],
        [0., np.cos(d), -np.sin(d)],
        [0., np.sin(d), np.cos(d)]
    ])
def rotx(d):
    d = d / 180. * np.pi

    return np.array([
        [np.cos(d), 0., np.sin(d)],
        [0., 1., 0.],
        [-np.sin(d), 0., np.cos(d)]
    ])
def roty(d):
    d = d / 180. * np.pi

    return np.array([
        [np.cos(d), -np.sin(d), 0.],
        [np.sin(d), np.cos(d), 0.],
        [0., 0., 1.]
    ])

def rot(m, x, y, z):
    rotmat = np.dot(rotz(z), np.dot(roty(y), rotx(x)))
    rotmatt = torch.tensor(rotmat, device=m.device)
    return meshfunc.rotate(m, rotmat=rotmatt, inplace=True)
```

```python
mesh = trimesh.load('/data/ugui0/kawana/ShapeNetLikePix3D/02691156/0b42484e10853efad0afe61e0c2035eb/model.off')
kalmesh = kal.rep.TriangleMesh.from_tensors(torch.tensor(mesh.vertices), torch.tensor(mesh.faces))
rot(kalmesh, 0, -90, -90)
mesh.vertices = kalmesh.vertices.to('cpu').numpy()
mesh.show()
```
